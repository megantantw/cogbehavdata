```python
import pandas as pd
import numpy as np

# read raw data - all cognitive data 
cd = pd.read_csv('totalcognitivedata.csv')
cd
```

```python
# Converted Naga's final filtered data into csv and open 
filtered_data = pd.read_csv('Naga_filtered_data.csv') # send Naga these 
filtered_data
```

```python
# merge both data frames together based on her identifiers 
merged_df = pd.merge(cd, filtered_data, on='Identifiers', how='right')
merged_df
```

```python
# Columns to check for periods ('.')
columns_to_filter = ['irep1 fluid_intelligence___reasoning__2_100027,fluid_intelligence_score', 'irep1 matrix_pattern_completion__2_501,number_of_puzzles_correctly_solved', 'irep1 reaction_time__2_100032,mean_time_to_correctly_identify_matches', 'irep1 symbol_digit_substitution__2_502,number_of_symbol_digit_matches_made_correctly', 'irep1 tower_rearranging__2_503,number_of_puzzles_correct', 'irep1 trail_making__2_505,duration_to_complete_alphanumeric_path_trail_#2', 'irep1 trail_making__2_505,duration_to_complete_numeric_path_trail_#1']

# Initialize an empty list to hold the index of rows to drop
rows_to_drop = []

# Iterate over the DataFrame
for index, row in merged_df.iterrows():
    # Check each specified column for a period in its string representation
    for column in columns_to_filter:
        if '.' in str(row[column]):
            # If a period is found, mark this row's index for exclusion
            rows_to_drop.append(index)
        elif 'NaN' in str(row[column]):
            rows_to_drop.append(index)
            break 

# Use the drop method to exclude the rows with periods in the specified columns
filtered_cognitive_data = merged_df.drop(rows_to_drop)
```
